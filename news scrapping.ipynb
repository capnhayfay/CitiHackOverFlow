```python
import pandas as pd
import numpy as np
import time
import requests 


from bs4 import BeautifulSoup

main_url = "https://finance.yahoo.com/quote/DB/news?p=DB"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
}

tic = time.perf_counter()
session = requests.Session()

website = session.get(main_url,headers = headers,timeout=20)
soup = BeautifulSoup(website.text, "html.parser")

links=[]
for record in soup.find_all("a",{"class":"js-content-viewer"}):
    links.append(record['href'])

session.close()

toc = time.perf_counter()
print("Time taken: " + str(toc-tic))

print(links)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

main_url = "https://finance.yahoo.com/quote/DB/news?p=DB"

driver = webdriver.Firefox()
driver.get(main_url)

start_time = end_time = time.perf_counter()

while (end_time - start_time < 10):
    driver.execute_script("window.scrollBy(0,2000)", "")
    end_time = time.perf_counter()

links = []
pages = driver.find_elements(By.CLASS_NAME,"js-content-viewer")

for page in pages:
    links.append(page.get_attribute("href"))
```
